# **Gathering Basic Insights from the data**

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/proccessed_data/train.csv')
df.head()
```

```python
df.info()
```

```python
df.hash_type.unique()
```

```python
len(df.hash_type.unique())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=df, x='hash_type', palette='coolwarm', hue='hash_type')
plt.xticks(rotation=90)
plt.show()
```

```python
df['hash_length'] = df['hash'].apply(lambda x: len(x))
```

```python
sns.barplot(data=df, x='hash_type', y='hash_length', palette='coolwarm', hue='hash_type')
plt.xticks(rotation=90)
plt.show()
```

```python
df.to_csv('../data/proccessed_data/train.csv', index=False)
```
